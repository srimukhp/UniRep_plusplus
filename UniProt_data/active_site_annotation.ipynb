Getting active site information from protein sequence and position of active site

```python
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_csv('active_site_annotation.tab', sep='\t', lineterminator='\n')
df_raw = df_raw.fillna('')
print 'Number of entries: ', len(df_raw)
df_raw = df_raw[df_raw['Length']<400] 
print 'Number of entries (len < 400): ', len(df_raw)
df_raw = df_raw.reset_index().drop('index',axis=1)
```

```python
df_active_site = pd.DataFrame(
    columns=['Entry','Entry name','Organism','Length','Active site num','Active site type','Active site residue', 'Sequence', 'UniRep'])
# df_active_site
```

```python
def get_act_num_type(act_str):
    act_type_found = False
    act_type = 0
    for act_site_type in act_site_type_codes.keys():
        if act_site_type in act_str:
            act_site = int(act_str.split(';')[0].strip())
#             print int(act_site.strip())
            act_type = act_site_type_codes[act_site_type]
            act_type_found = True
    if act_type_found:
        return act_site, act_type
    else:
        return None, None

def get_aa_from_seq(seq, aa_num):
    aa_names = []
    lst_seq = list(seq)
    return lst_seq[aa_num-1]

act_site_type_codes = {'Charge relay system':1,
                       'Nucleophile':2,
                       'Electrophile':3,
                       'Proton donor':4,
                       'Proton acceptor':5}
```

```python
for idx, row in df_raw.iterrows():
#     print get_act_num_type(act_strs[1])
    act_strs = row['Active site']
    act_strs = act_strs.split('ACT_SITE')
    act_strs = [x for x in act_strs if len(x)>1]
    if (idx % 1000 ==0): print 'Row num: ',idx
    for act_str in act_strs:
        act_site, act_type =  get_act_num_type(act_str)
        if not (act_site == None or act_type == None):
            pd_series = df_raw.iloc[idx][['Entry','Entry name','Organism','Length','Sequence']]
#             print pd_series
            add_series = pd.Series([act_site,act_type],index=['Active site num','Active site type'])
            pd_series =  pd_series.append(add_series)
#             print pd_series
            df_active_site = df_active_site.append(pd_series, ignore_index=True)
```

```python
for idx, row in df_active_site.iterrows():
    if (idx%1000==0): print idx
    num = df_active_site.iloc[idx]['Active site num']
    seq = df_active_site.iloc[idx]['Sequence']
    df_active_site.iloc[idx]['Active site residue'] = get_aa_from_seq(seq, num)
```
