Getting active site + neighbor information from protein sequence and active site position

```python
import pandas as pd
import numpy as np
```

```python
df_act = pd.read_excel('active_site_annotation.xlsx',header=[0],index_col=[0])
df_act['Active site neighbor'] = np.nan
```

```python
df_act.head()
```

```python
def get_act_site_neighbor(seq, act_num):
    return seq[act_num-2:act_num+1]
```

```python
for idx, row in df_act.iterrows():
    if (idx%1000==0): print idx
    num = df_act.iloc[idx]['Active site num']
    seq = df_act.iloc[idx]['Sequence']
    df_act.ix[idx, 'Active site neighbor'] = get_act_site_neighbor(seq, num)
```

```python
df_act.head()
```

```python
df_groupby_entry = df_act.groupby(['Entry','Active site neighbor'],axis=0).agg({"Active site type": "nunique"})
len_nonunique_rep = len(df_groupby_entry[df_groupby_entry['Active site type'] > 1])
```
