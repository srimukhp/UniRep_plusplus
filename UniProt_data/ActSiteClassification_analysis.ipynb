```python
df_active_site = pd.read_excel('active_site_annotation.xlsx',header=[0],index_col=[0])
df_active_site.head()
```

```python
# df_active_site.groupby(by=['Active site residue','Active site type'],axis=0)['Length'].count()
df_grouped = df_active_site.groupby(by=['Active site residue','Active site type'],axis=0)['Length'].count()
df_percent = (df_grouped / df_grouped.sum(level=0) )* 100.
df_percent = df_percent[['C','D','E','H','K','R','S','T','Y']]
df_percent = df_percent.to_frame()
# df_percent.reset_index()
```

```python
aa_list = df_percent.index.get_level_values(0).unique().tolist()
index = pd.MultiIndex.from_product([aa_list, np.arange(1,6)])
new_df = pd.DataFrame(0. ,index=index, columns = ['Frequency'])
new_df=new_df.reset_index()
new_df.head()
```

```python
merged_df = pd.merge(new_df, df_percent, left_on=['level_0','level_1'], right_on=['Active site residue','Active site type'],how='outer')
```

```python
merged_df = merged_df.drop('Frequency',axis=1).fillna(0.)
merged_df.head()
```

```python
actsite_1 =  merged_df[merged_df['level_1'] == 1]['Length'].tolist()
actsite_2 =  merged_df[merged_df['level_1'] == 2]['Length'].tolist()
actsite_3 =  merged_df[merged_df['level_1'] == 3]['Length'].tolist()
actsite_4 =  merged_df[merged_df['level_1'] == 4]['Length'].tolist()
actsite_5 =  merged_df[merged_df['level_1'] == 5]['Length'].tolist()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
label_fontsize = 16

def autolabel(rects, n_round=1, fontsize=16):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{0:.{1}f}%'.format(height, n_round),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points", fontsize=fontsize,
                    ha='center', va='bottom')
```

```python
act_site_type_codes = {'Charge relay system':1,
                       'Nucleophile':2,
                       'Electrophile':3,
                       'Proton donor':4,
                       'Proton acceptor':5}
labels = ['cysteine','aspartic acid','glutamic acid','histidine','lysine','arginine','serine','threonine','tyrosine']
x = np.arange(len(labels))  # the label locations
width = 0.15  # the width of the bars
fig, ax = plt.subplots(figsize=(14,10))
rects1 = ax.bar(x - 2*width, actsite_1, width, label='Charge relay system',edgecolor='black')
rects2 = ax.bar(x - width, actsite_2, width, label='Nucleophile',edgecolor='black')
rects3 = ax.bar(x , actsite_3, width, label='Electrophile',edgecolor='black')
rects4 = ax.bar(x + width, actsite_4, width, label='Proton donor',edgecolor='black')
rects5 = ax.bar(x + 2* width, actsite_5, width, label='Proton acceptor',edgecolor='black')

ax.set_xlabel('Amino acids', fontsize=label_fontsize+6)
ax.set_ylabel('Percentage of sites', fontsize=label_fontsize+6)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.tick_params(axis="y", labelsize=label_fontsize)
# ax.set_yticklabels( fontsize=label_fontsize)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=label_fontsize, rotation=90)
ax.legend(fontsize=label_fontsize)
# plt.show()
plt.savefig('residue_function_dist.pdf',dpi=400,bbox_inches='tight')
plt.close()
```
